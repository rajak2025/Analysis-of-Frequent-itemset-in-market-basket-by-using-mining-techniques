```python
#Importing the data
import os
print(os.listdir("../input"))
#We can see that the dataset has data of Data,Time,Transaction and the item sold at the bakery.
df = pd.read_csv('../input/BreadBasket_DMS.csv')

df.head(), df.info()
```

```python
#Transforming all item names to lower case

df['Item'] = df['Item'].str.lower()
```

```python
#Inspecting the data

x = df['Item'] == "none"
print(x.value_counts())
```

```python
#Dropping all none values
df = df.drop(df[df.Item == 'none'].index)
```

```python
print(df)
```

```python
import numpy as np
list1 = [2,3]
array2 = np.array(df)
newArray = array2[:, list1]
newList = newArray.tolist()
```

```python
print(newList)
```

```python
d = {}

for values in newList :
    k = values[0]
    v = values[1]
    d.setdefault(k,[]).append(v)
```
